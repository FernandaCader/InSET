# Análisis Exploratorio de Datos de usuarios.csv

```python
import pandas as pd

# 2.1 Cargar el archivo en un DataFrame Pandas
df = pd.read_csv("usuarios.csv")

# 2.2 Obtener el name y email de todos los usuarios
name_email = df[['name', 'email']]
```

```python
# 2.3 Obtener el name de todos los usuarios cuyo gender es female
female = df[df['gender'] == 'female']['name']
```

```python
# 2.4 Obtener el name de todos los usuarios cuyo gender es male y tengan una age mayor a 50
male_mayor_50 = df[(df['gender'] == 'male') & (df['age'] > 50)]['name']
```

```python
# 2.5 Obtener todos los usuarios cuya age sea mayor a 20
mayor_20 = df[df['age'] > 20]
```

```python
# 2.6 Obtener el username y email electrónico de los usuarios cuya age se encuentre en el rango 50 a 60
age_50_60 = df[(df['age'] >= 50) & (df['age'] <= 60)][['username', 'email']]
```

```python
# 2.7 Obtener el username de todos los usuarios cuyo email no termine con @example.com
email_no_example = df[~df['email'].str.endswith('@example.com')]['username']
```

```python
# 2.8 Obtener el username de todos los usuarios cuyo country sea Germany, Finland o Canada
paises = ['Germany', 'Finland', 'Canada']
paises_users = df[df['country'].isin(paises)]['username']
```

```python
# 2.9 Obtener el name y email electrónico de todos los usuarios de gender female que sean del country Germany
female_Germany = df[(df['gender'] == 'female') & (df['country'] == 'Germany')][['name', 'email']]
```

```python
# 2.10 Obtener el promedio de age de todos los usuarios de gender female de Canada mayores de 20 años
promedio_age_female_canada = df[(df['gender'] == 'female') & (df['country'] == 'Canada') & (df['age'] > 20)]['age'].mean()
```

```python
# 2.11 Conocer la cantidad de usuarios que residen en Finland
usuarios_Finland = len(df[df['country'] == 'Finland'])
```

```python
# 2.12 Mostrar en consola la cantidad de hombres y mujeres en el dataframe
cantidad_hombres = len(df[df['gender'] == 'male'])
cantidad_mujeres = len(df[df['gender'] == 'female'])
print(f"Cantidad de Hombres: {cantidad_hombres}")
print(f"Cantidad de Mujeres: {cantidad_mujeres}")
```

```python
# 2.13 Mostrar en consola el country con más mujeres
pais_mas_mujeres = df[df['gender'] == 'female']['country'].value_counts().idxmax()
print(f"country con más mujeres: {pais_mas_mujeres}")
```

```python
# 2.14 Obtener los 3 countryes con más usuarios
paises_top_3 = df['country'].value_counts().head(3)
```

```python
# 2.15 Mostrar en consola el name de todos los usuarios cuya age se encuentre en el rango de 10 a 20 y 40 a 70
agees_rango = df[df['age'].between(10, 20) | df['age'].between(40, 70)]['name']
```

```python
# 2.16 Mostrar en pantalla el name y email electrónico del usuario más joven en Canada
usuario_mas_joven_canada = df[(df['country'] == 'Canada') & (df['age'] == df[df['country'] == 'Canada']['age'].min())][['name', 'email']]
```

```python
# 2.17 Mostrar en pantalla el name y email electrónico del usuario más viejo en Canada
usuario_mas_viejo_canada = df[(df['country'] == 'Canada') & (df['age'] == df[df['country'] == 'Canada']['age'].max())][['name', 'email']]
```

```python
# 2.18 Listar en consola los 3 countryes con menor cantidad de usuarios
paises_menos_usuarios = df['country'].value_counts().tail(3)
```

```python
# 2.19 Obtener el country con mayor cantidad de usuarios cuya age sea mayor a 50
pais_mayor_age_50 = df[df['age'] > 50]['country'].value_counts().idxmax()
```

```python
# 2.20 Obtener el country con mayor promedio de age
pais_mayor_promedio_age = df.groupby('country')['age'].mean().idxmax()
```

```python
# 2.21 Mostrar en consola el country con más hombres
pais_mas_hombres = df[df['gender'] == 'male']['country'].value_counts().idxmax()
```

```python
# 2.22 Mostrar en consola el name, username y age de todos los usuarios cuya age es mayor a 10 y no sean del country Mexico, Brazil y Canada
usuarios_age_mayor_10 = df[(df['age'] > 10) & (~df['country'].isin(['Mexico', 'Brazil', 'Canada']))][['name', 'username', 'age']]
```

```python
# 2.23 Mostrar en consola el postcode de todos los usuarios de Mexico
codigos_postales_mexico = df[df['country'] == 'Mexico']['postcode']
```

```python
# 2.24 Obtener la age que más se repite en el DataFrame
age_mas_repetida = df['age'].mode().values[0]
```

```python
# 2.25 Obtener la age que menos se repite en el DataFrame
age_menos_repetida = df['age'].value_counts().tail(1).index[0]
```

```python
# Imprimir resultados
print(name_email)
print(female)
print(male_mayor_50)
print(mayor_20)
print(age_50_60)
print(email_no_example)
print(paises_users)
print(female_Germany)
print(promedio_age_female_canada)
print(usuarios_Finland)
print(cantidad_hombres)
print(cantidad_mujeres)
print(pais_mas_mujeres)
print(paises_top_3)
print(agees_rango)
print(usuario_mas_joven_canada)
print(usuario_mas_viejo_canada)
print(paises_menos_usuarios)
print(pais_mayor_age_50)
print(pais_mayor_promedio_age)
print(pais_mas_hombres)
print(usuarios_age_mayor_10)
print(codigos_postales_mexico)
print(age_mas_repetida)
print(age_menos_repetida)
```

# Gráficos

2.5 Todos los usuarios cuya edad sea mayor a 20

```python
import matplotlib.pyplot as plt

# Filtrar usuarios cuya edad sea mayor a 20
usuarios_mayor_20 = df[df['age'] > 20]

# Crear un gráfico de caja y bigotes
plt.figure(figsize=(8, 6))
plt.boxplot(usuarios_mayor_20['age'], vert=False)  # vert=False para mostrar la caja horizontalmente
plt.xlabel('Edad')
plt.title('Distribución de Edades de Usuarios Mayores de 20')
plt.show()
```

2.10: Histograma de edades de mujeres de Canadá mayores de 20 años

```python
# Filtrar datos
filtro = (df['gender'] == 'female') & (df['country'] == 'Canada') & (df['age'] > 20)
edades_femenino_canada = df[filtro]['age']

# Crear un histograma de edades
plt.hist(edades_femenino_canada, bins=10, edgecolor='black')
plt.xlabel('Edad')
plt.ylabel('Cantidad de Usuarios')
plt.title('Histograma de Edades de Mujeres en Canadá (Mayores de 20 años)')
plt.show()
```

2.14: Gráfico de barras de los 3 países con más usuarios

```python
# Obtener los 3 países con más usuarios
top_3_countries = df['country'].value_counts().head(3)

# Crear un gráfico de barras
top_3_countries.plot(kind='bar')
plt.xlabel('País')
plt.ylabel('Cantidad de Usuarios')
plt.title('Top 3 Países con Más Usuarios')
plt.show()
```

2.19: Gráfico de pastel del país con la mayor cantidad de usuarios mayores de 50 años

```python
# Filtrar datos
filtro = df['age'] > 50

# Obtener la cantidad de usuarios mayores de 50 años por país
conteo_paises = df[filtro]['country'].value_counts()

# Crear un gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(conteo_paises.index, conteo_paises.values)
plt.xlabel('País')
plt.ylabel('Cantidad de Usuarios Mayores de 50 años')
plt.title('Cantidad de Usuarios Mayores de 50 años por País')
plt.xticks(rotation=45)  # Rotar etiquetas del eje x para una mejor legibilidad
plt.tight_layout()  # Ajustar el diseño para evitar recorte de etiquetas
plt.show()
```

2.22: Gráfico de barras para usuarios cuya edad es mayor a 10 y no son de México, Brasil o Canadá

```python
# Filtrar datos
filtro = (df['age'] > 10) & (~df['country'].isin(['Mexico', 'Brazil', 'Canada']))
usuarios_edad_mayor_10 = df[filtro]

# Obtener una cuenta de usuarios por país
conteo_paises = usuarios_edad_mayor_10['country'].value_counts()

# Crear un gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(conteo_paises.index, conteo_paises.values)
plt.xlabel('País')
plt.ylabel('Cantidad de Usuarios')
plt.title('Usuarios con Edad Mayor a 10 (Excluyendo México, Brasil y Canadá) por País')
plt.xticks(rotation=45)  # Rotar etiquetas del eje x para una mejor legibilidad
plt.tight_layout()  # Ajustar el diseño para evitar recorte de etiquetas
plt.show()
```
