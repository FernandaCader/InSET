```python
import cv2
import matplotlib.pyplot as plt

# Cargar la imagen
image_path = 'rostros.jpg'
image = cv2.imread(image_path)

# Convertir la imagen a escala de grises
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Mostrar la imagen original
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')
plt.axis('off')
plt.show()
```

```python
# Cargar el clasificador en cascada para la detección de rostros
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
# Detectar rostros en la imagen
faces = face_cascade.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

# Dibujar cuadros delimitadores alrededor de los rostros detectados
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 2)

# Mostrar la imagen con los rostros detectados
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Rostros Detectados')
plt.axis('off')
plt.show()
```

```python
# Detectar rostros con parámetros ajustados
adjusted_faces = face_cascade.detectMultiScale(gray_image, scaleFactor=1.05, minNeighbors=3, minSize=(20, 20))

# Dibujar cuadros delimitadores alrededor de los rostros detectados con parámetros ajustados
for (x, y, w, h) in adjusted_faces:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Mostrar la imagen con los rostros detectados utilizando parámetros ajustados
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Rostros Detectados con Parámetros Ajustados')
plt.axis('off')
plt.show()
```

# Informe
## Parámetros Utilizados en detectMultiScale

- scaleFactor: Este parámetro especifica cuánto se reduce el tamaño de la imagen en cada escala de la imagen. Un valor de 1.1 significa que la imagen se reduce en un 10% en cada escala.
- minNeighbors: Este parámetro define cuántos vecinos cada rectángulo candidato debe tener para ser retenido. Un número más alto significa menos detecciones pero de mayor calidad.
- minSize: Este parámetro especifica el tamaño mínimo de la ventana para ser considerada como un rostro. En este caso, se ha establecido en (30, 30) píxeles.

## Análisis de los Resultados Obtenidos
En general, los parámetros utilizados son adecuados para la detección básica de rostros en imágenes. Sin embargo, dependiendo de la calidad y las características de la imagen, es posible que se necesiten ajustes:

- Número de detecciones: Si la imagen contiene varios rostros y todos fueron detectados correctamente, los parámetros son adecuados.
- Falsos positivos: Si se detectaron objetos que no son rostros, es posible que sea necesario aumentar minNeighbors.
- Falsos negativos: Si algunos rostros no fueron detectados, podría ser útil reducir minNeighbors o ajustar scaleFactor.

## Posibles Mejoras y Ajustes en los Parámetros
- scaleFactor: Reducir este valor (por ejemplo, a 1.05) puede ayudar a detectar rostros de diferentes tamaños, pero aumentará el tiempo de cómputo.
- minNeighbors: Ajustar este valor puede ayudar a equilibrar entre falsos positivos y falsos negativos. Un valor más bajo puede aumentar las detecciones, pero también los falsos positivos.
- minSize: Ajustar el tamaño mínimo de la ventana puede ser útil para detectar rostros más pequeños o ignorar elementos que no son rostros.
