```python
import cv2
import matplotlib.pyplot as plt

# Cargar la imagen
image_path = 'rostros.jpg'
image = cv2.imread(image_path)

# Convertir la imagen a escala de grises
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Mostrar la imagen original
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')
plt.axis('off')
plt.show()
```

```python
# Cargar el clasificador en cascada para la deteccin de rostros
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
# Detectar rostros en la imagen
faces = face_cascade.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

# Dibujar cuadros delimitadores alrededor de los rostros detectados
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 2)

# Mostrar la imagen con los rostros detectados
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Rostros Detectados')
plt.axis('off')
plt.show()
```

```python
# Detectar rostros con parmetros ajustados
adjusted_faces = face_cascade.detectMultiScale(gray_image, scaleFactor=1.05, minNeighbors=3, minSize=(20, 20))

# Dibujar cuadros delimitadores alrededor de los rostros detectados con parmetros ajustados
for (x, y, w, h) in adjusted_faces:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Mostrar la imagen con los rostros detectados utilizando parmetros ajustados
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Rostros Detectados con Parmetros Ajustados')
plt.axis('off')
plt.show()
```

# Informe
## Parmetros Utilizados en detectMultiScale

- scaleFactor: Este parmetro especifica cunto se reduce el tamao de la imagen en cada escala de la imagen. Un valor de 1.1 significa que la imagen se reduce en un 10% en cada escala.
- minNeighbors: Este parmetro define cuntos vecinos cada rectngulo candidato debe tener para ser retenido. Un nmero ms alto significa menos detecciones pero de mayor calidad.
- minSize: Este parmetro especifica el tamao mnimo de la ventana para ser considerada como un rostro. En este caso, se ha establecido en (30, 30) pxeles.

## Anlisis de los Resultados Obtenidos
En general, los parmetros utilizados son adecuados para la deteccin bsica de rostros en imgenes. Sin embargo, dependiendo de la calidad y las caractersticas de la imagen, es posible que se necesiten ajustes:

- Nmero de detecciones: Si la imagen contiene varios rostros y todos fueron detectados correctamente, los parmetros son adecuados.
- Falsos positivos: Si se detectaron objetos que no son rostros, es posible que sea necesario aumentar minNeighbors.
- Falsos negativos: Si algunos rostros no fueron detectados, podra ser til reducir minNeighbors o ajustar scaleFactor.

## Posibles Mejoras y Ajustes en los Parmetros
- scaleFactor: Reducir este valor (por ejemplo, a 1.05) puede ayudar a detectar rostros de diferentes tamaos, pero aumentar el tiempo de cmputo.
- minNeighbors: Ajustar este valor puede ayudar a equilibrar entre falsos positivos y falsos negativos. Un valor ms bajo puede aumentar las detecciones, pero tambin los falsos positivos.
- minSize: Ajustar el tamao mnimo de la ventana puede ser til para detectar rostros ms pequeos o ignorar elementos que no son rostros.
